```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting visual style for all plots
sns.set(style="whitegrid")
```

```python
# Reading the Airbnb NYC dataset
df = pd.read_csv("AB_NYC_2019.csv")
```

# 1. Creating a stacked bar plot to show average price by neighbourhood group and room type

```python
# Grouping data by neighbourhood group and room type, then calculating average price
grouped_data = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().unstack()
```

```python
# Creating the stacked bar plot
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title("Stacked Bar Plot of Average Price by Neighbourhood Group and Room Type")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.legend(title="Room Type")
plt.tight_layout()
plt.show()
```

# 2. Creating a box plot to show price distribution by room type

```python
# Filtering out prices higher than 500 to remove extreme values
filtered_df = df[df['price'] < 500]
```

```python
# Creating the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_df, x='room_type', y='price', palette='Set2')
plt.title("Box Plot of Price Distribution by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price")
plt.tight_layout()
plt.show()
```

# 3. Creating a bar plot to show top 10 neighbourhoods by number of listings

```python
# Finding the 10 neighbourhoods with the highest number of listings
top_neighbourhoods = df['neighbourhood'].value_counts().nlargest(10)
```

```python
# Creating the horizontal bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=top_neighbourhoods.values, y=top_neighbourhoods.index, palette="viridis")
plt.title("Top 10 Neighbourhoods by Number of Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Neighbourhood")
plt.tight_layout()
plt.show()
```

# 4. Creating a scatter plot to show price vs number of reviews

```python
# Filtering out price outliers above 1000 for clarity
scatter_df = df[df['price'] < 1000]
```

```python
# Creating the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=scatter_df, x='number_of_reviews', y='price', hue='room_type', alpha=0.6)
plt.title("Scatter Plot of Price vs Number of Reviews")
plt.xlabel("Number of Reviews")
plt.ylabel("Price")
plt.legend(title="Room Type")
plt.tight_layout()
plt.show()
```
